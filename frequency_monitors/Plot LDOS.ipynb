```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
def read_ldos(fname):
    data = genfromtxt(fname, delimiter=",", dtype='str');
    data = data[:,1:]
    data = char.replace(data, 'i', 'j').astype(float64);

    freq = data[:, 0]
    ldos = data[:, 1]
    
    return (freq, ldos)
```

```python
(freq, ldos) = read_ldos('ldos_ring')
plt.plot(freq, ldos)
plt.xlabel('Frequency')
plt.ylabel('LDOS')
plt.show()
```
